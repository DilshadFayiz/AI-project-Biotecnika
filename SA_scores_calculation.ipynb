# Task
Calculate the Synthetic Accessibility (SA) scores for the molecules in the file "/content/ADMET_novel_anticancer_candidates.csv" and display the molecules along with their SA scores.

```python
import pandas as pd
from rdkit import Chem
from rdkit.Contrib.SA_Score.sascorer import calculateScore as calculate_sa_score_rdkit

# Load the data
df = pd.read_csv('/content/ADMET_novel_anticancer_candidates.csv')

# Define the SA score calculation function
def calculate_sa_score(smiles):
    """Calculates the Synthetic Accessibility (SA) score for a given SMILES string."""
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None
    return calculate_sa_score_rdkit(mol)

# Calculate SA scores and add to DataFrame
df['SA_Score'] = df['SMILES'].apply(calculate_sa_score)

# Display molecules with lowest and highest SA scores
molecule_with_lowest_sa = df.sort_values(by='SA_Score', ascending=True).iloc[0]
molecule_with_highest_sa = df.sort_values(by='SA_Score', ascending=False).iloc[0]
display(molecule_with_lowest_sa[['MOLECULE_ID', 'SA_Score']])
display(molecule_with_highest_sa[['MOLECULE_ID', 'SA_Score']])
```
