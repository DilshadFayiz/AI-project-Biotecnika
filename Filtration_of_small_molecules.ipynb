```python
import pandas as pd

# Load the data
df = pd.read_csv('/content/AKT1_inhibitors.csv')

# Filter Standard Value <= 1000
filtered_df = df[df['Standard Value'] <= 1000]

# Remove rows with Standard Relation '>' or '<'
filtered_df = filtered_df[~filtered_df['Standard Relation'].isin(['>', '<'])]

# Keep only specified columns and filter Molecular Weight <= 550
filtered_df = filtered_df[['Molecule Name', 'Molecular Weight', 'Smiles', 'Standard Value', 'Target Name']]
filtered_df = filtered_df[filtered_df['Molecular Weight'] <= 550]

# Remove rows with empty 'Target Name'
filtered_df = filtered_df.dropna(subset=['Target Name'])

# Keep only rows where 'Target Name' is 'RAC-alpha serine/threonine-protein kinase'
filtered_df = filtered_df[filtered_df['Target Name'] == 'RAC-alpha serine/threonine-protein kinase']

display(filtered_df.head())
```

```python
filtered_df.to_csv('filtered_akt1_inhibitors.csv', index=False)
```
